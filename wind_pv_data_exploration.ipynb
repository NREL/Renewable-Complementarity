```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import colors
```

### Read in hourly data
#### These data describe mean annual complementarity findings based on modeled hourly capacity factors spanning 2007-2013

```python
# Note that Pearson data include both multi-year mean annual and mean monthly correlations
pearson_hourly = pd.read_csv('data/wind-pv/hourly/pearson_my.csv')
```

```python
kendall_hourly = pd.read_csv('data/wind-pv/hourly/kendall_my.csv')
```

```python
spearman_hourly = pd.read_csv('data/wind-pv/hourly/spearman_my.csv')
```

### Read in daily data
#### The data describe mean annual complementarity findings based on modeled daily average capacity factors (from hourly data) spanning 2007-2013
#### The stability coefficient is a daily metric based on hourly data

```python
pearson_daily = pd.read_csv('data/wind-pv/daily/pearson_my.csv')
```

```python
stability_coefficient = pd.read_csv('data/wind-pv/daily/cstab_my.csv')
```

### Prep visualizations

```python
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
```

```python
# Center the bivariate color ramp at 0 so that blues indicate negative correlations (i.e., complementarity),
# and reds indicate positive correlations (i.e., lack of complementarity)
divnorm=colors.TwoSlopeNorm(vmin=-1., vcenter=0., vmax=1.)
```

### Visualize mean annual Pearson Correlation for wind-PV hybrids (hourly)

```python
fig, ax = plt.subplots(figsize=(10,6))

pearson_hourly.sample(30000).plot.scatter(ax=ax, x='longitude',y='latitude',c='pearson_my_mean',cmap='coolwarm',norm=divnorm)
plt.title('Mean annual Pearson')
f = plt.gcf()
cax = f.get_axes()[1]
cax.set_ylabel('Pearson correlation coeff.')
plt.show()
```

### Visualize monthly variation in Pearson Correlation for wind-PV hybrids (hourly)

```python
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

for month in months:
    fig, ax = plt.subplots(figsize=(5,3))
    col = '{}_pearson_mean'.format(month)
    pearson_hourly.sample(30000).plot.scatter(ax=ax, x='longitude',y='latitude',c=col,cmap='coolwarm',norm=divnorm)
    plt.title(col)
    ax.set_xlabel('longitude')
    f = plt.gcf()
    cax = f.get_axes()[1]
    cax.set_ylabel('Pearson correlation coeff.')
    plt.show()
```

### Visualize mean annual daily Pearson Correlation for wind-PV hybrids

```python
fig, ax = plt.subplots(figsize=(10,6))

pearson_daily.sample(30000).plot.scatter(ax=ax, x='longitude',y='latitude',c='pearson_my_mean',cmap='coolwarm',norm=divnorm)
plt.title('Mean annual Pearson (daily correlation)')
f = plt.gcf()
cax = f.get_axes()[1]
cax.set_ylabel('Pearson correlation coeff.')
plt.show()
```

### Visualize monthly variation in daily Pearson Correlation for wind-PV hybrids

```python
for month in months:
    fig, ax = plt.subplots(figsize=(5,3))
    col = '{}_pearson_mean'.format(month)
    pearson_daily.sample(30000).plot.scatter(ax=ax, x='longitude',y='latitude',c=col,cmap='coolwarm',norm=divnorm)
    plt.title('{}, daily'.format(col))
    ax.set_xlabel('longitude')
    f = plt.gcf()
    cax = f.get_axes()[1]
    cax.set_ylabel('Pearson correlation coeff.')
    plt.show()
```

### Visualize mean annual stability coefficient for wind-PV hybrids

```python
fig, ax = plt.subplots(figsize=(10,6))

stability_coefficient.sample(30000).plot.scatter(ax=ax, x='longitude',y='latitude',c='stability_my_mean',cmap='Blues')
plt.title('Mean annual Pearson (daily correlation)')
f = plt.gcf()
cax = f.get_axes()[1]
cax.set_ylabel('Stability coefficient')
plt.show()
```

### Visualize monthly variation in stability coefficient for wind-PV hybrids

```python
for month in months:
    fig, ax = plt.subplots(figsize=(5,3))
    col = '{}_stability_mean'.format(month)
    stability_coefficient.sample(30000).plot.scatter(ax=ax, x='longitude',y='latitude',c=col,cmap='Blues')
    plt.title('{}, daily'.format(col))
    ax.set_xlabel('longitude')
    f = plt.gcf()
    cax = f.get_axes()[1]
    cax.set_ylabel('Stability coefficient')
    plt.show()
```
